```python
base_address = 'http://ois2.ut.ee/api'
```

```python
def get_all_courses():
    courses = []
    parameters = {
        'start': 1,
        'take': 300,
        }
    received = 300
    while parameters['take'] == received:
        response = requests.get(f'{base_address}/courses', parameters)
        received = len(response.json())
        courses += response.json()
        parameters['start'] += parameters['take']
        print(f'Collected {len(courses)} courses')
    
    return courses

courses = get_all_courses()    
```

## Latest course endpoint data gathering / cleaning

```python
def get_course_latest_details(course):
    return requests.post(f'{base_address}/courses/course-version/{course["latest_version_uuid"]}').json()

# get_course_latest_details(nlp_course)
```

```python
def get_latest_course_info(courses):
    latest_info = []
    for course in tqdm(courses):
        if 'latest_version_uuid' in course:
            latest_info.append(get_course_latest_details(course))
        
    return latest_info

latest_course_info = get_latest_course_info(courses)
```

```python
def flatten_col(df, col):
    return df.join(pd.json_normalize(df[col]).add_prefix(f'{col}_')).drop(col, axis='columns')

def array_contains_english_keys(arr):
        for obj in arr:
            if not 'en' in obj.keys():
                return False

        return True

def flatten_multilingual_array(df, col):
    nan_value_rows = df.loc[df[col].apply(lambda x: isinstance(x, float))]
    nan_mask = df['uuid'].isin(nan_value_rows['uuid'])
    empty_lists = pd.Series([[]] * len(nan_value_rows), index=nan_value_rows.index)
    df.loc[nan_mask, col] = empty_lists
    
    non_en_rows = df.loc[df[col].apply(lambda arr: not array_contains_english_keys(arr))]
    non_en_mask = df['uuid'].isin(non_en_rows['uuid'])
    empty_lists = pd.Series([[]] * len(non_en_rows), index=non_en_rows.index)
    df.loc[non_en_mask, col] = empty_lists
    
    df[col] = df[col].apply(lambda arr: [obj['en'] for obj in arr])
    
    return df
```

```python
def clean_latest_course_data(data_df):
    latest_course_data_df = data_df.copy()
    
    latest_course_data_df = flatten_col(latest_course_data_df, 'title')
    latest_course_data_df = flatten_col(latest_course_data_df, 'general')
    latest_course_data_df = flatten_col(latest_course_data_df, 'additional_info')
    latest_course_data_df = flatten_col(latest_course_data_df, 'overview')
    latest_course_data_df = flatten_col(latest_course_data_df, 'target')
    latest_course_data_df = flatten_col(latest_course_data_df, 'grading')
    latest_course_data_df = flatten_col(latest_course_data_df, 'resources')
    latest_course_data_df = flatten_col(latest_course_data_df, 'participants')
    latest_course_data_df = flatten_col(latest_course_data_df, 'schedule')
    latest_course_data_df = flatten_col(latest_course_data_df, 'registration_info')
    
    latest_course_data_df = flatten_multilingual_array(latest_course_data_df, 'overview_objectives')
    latest_course_data_df = flatten_multilingual_array(latest_course_data_df, 'overview_learning_outcomes')
    
    # map grading_independent_work_assesments
    def map_independent_work_assesments(assessments):
        if type(assessments) != list:
            return []
        
        return [{
            'work_type_code': item['work_type']['code'],
            'work_type_name': item['work_type']['en'],
            'description': item['description']['en'] if 'en' in item['description'] else ''
            } for item in assessments]
        
    latest_course_data_df['grading_independent_work_assessments'] = latest_course_data_df['grading_independent_work_assessments'].apply(lambda x : map_independent_work_assesments(x))
    
    def map_general_input_languages(languages):
        if type(languages) != list:
            return []
        
        return [{
            'language_code': item['code'],
            'language_name': item['en'],
            } for item in languages]
        
    latest_course_data_df['general_input_languages'] = latest_course_data_df['general_input_languages'].apply(lambda x : map_general_input_languages(x))
    
    def map_general_structural_unit_share(structural_units):
        if type(structural_units) != list:
            return []
        
        return [{
            'code': item['structural_unit_code'],
            'name': item['structural_unit_name']['en'],
            'course_share_percent': item['course_share_percent'],
            'coordinating': item['coordinating']
        } for item in structural_units]
        
    latest_course_data_df['general_structural_unit_shares'] = latest_course_data_df['general_structural_unit_shares'].apply(lambda x : map_general_structural_unit_share(x))
    
    def map_study_levels(study_levels):
        if type(study_levels) != list:
            return []
        
        return[{
            'code': item['code'],
            'title': item['en']
        } for item in study_levels]
    
    latest_course_data_df['additional_info_study_levels'] = latest_course_data_df['additional_info_study_levels'].apply(lambda x : map_study_levels(x))
    
    def map_languages(languages):
        if type(languages) != list:
            return []
        
        return [{
            'code': item['code'],
            'name': item['en'],
            } for item in languages]
    
    latest_course_data_df['overview_study_languages'] = latest_course_data_df['overview_study_languages'].apply(lambda x : map_languages(x))
    latest_course_data_df['overview_other_languages'] = latest_course_data_df['overview_other_languages'].apply(lambda x : map_languages(x))
    
    columns_to_drop = ['last_update', 'credit_type', 'confirmation_process', 'credits',
                       'state', 'title_et', 'title_vro', 'title_es',
                       'general_year.code', 'general_year.et', 'general_type.et',
                       'additional_info_hours.colloquium',
                       'additional_info_hours.web_learning',
                       'additional_info_language_training.language.et',
                       'additional_info_language_training.start_level.et',
                       'additional_info_language_training.start_level.en',
                       'additional_info_language_training.end_level.et',
                       'additional_info_language_training.end_level.en',
                       'target_year.et', 'target_semester.et',
                       'target_semester.en', 'target_study_type.et',
                       'target_study_type.code', 'target_language.code',
                       'target_language.et', 'target_part.code', 'target_part.en',
                       'target_faculty.name.et', 'target_faculty.academic',
                       'target_faculty.supports_continuous_learning',
                       'target_faculty.parent_code',
                       'target_course_main_structural_unit.name.et',
                       'target_course_main_structural_unit.academic',
                       'target_label.et',
                       'target_course_main_structural_unit.is_expired',
                       'target_faculty.is_expired',
                       'grading_assessment_scale.et',
                       'grading_grade_conditions.A.et',
                       'grading_grade_conditions.B.et',
                       'grading_grade_conditions.C.et',
                       'grading_grade_conditions.D.et',
                       'grading_grade_conditions.F.et',
                       'resources_web_based.et',
                       'resources_web_based.code',
                       'resources_library_package',
                       'participants_study_groups',
                       'registration_info_audience.et'
                       ]
    # parent_uuid, parent_code, parent_credits
    latest_course_data_df.drop(columns=columns_to_drop, inplace=True)
    
    return latest_course_data_df

cleaned_latest_course_info_df = clean_latest_course_data(latest_course_info_df)
```
